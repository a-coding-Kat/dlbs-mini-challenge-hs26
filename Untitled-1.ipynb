```python
import os
import random
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

file_path = './data/test/santa'

# --- Modular Helper Functions (Recap) ---

def get_and_process_image(file_path):
    """Opens an image and returns the PIL Image object and its NumPy array."""
    try:
        img = Image.open(file_path)
        img_array = np.array(img)
        return img, img_array
    except Exception as e:
        print(f"Error opening image at {file_path}: {e}")
        return None, None

def display_image_with_pixels(img, filename, plot_position):
    """Displays the image and prints its dimensions."""
    width, height = img.size
    total_pixels = width * height
    
    plt.subplot(plot_position)
    plt.imshow(img)
    plt.title(f'Image: {filename}')
    plt.axis('off')
    plt.xlabel(f'{width}x{height} pixels | Total: {total_pixels:,}')

def plot_color_histogram(img_array, plot_position):
    """Calculates and plots the color histogram (RGB distribution)."""
    plt.subplot(plot_position)
    plt.title("Color Histogram")
    plt.xlabel("Pixel Intensity Value (0-255)")
    plt.ylabel("Frequency (Pixel Count)")
    
    color_channels = ('r', 'g', 'b')
    channel_names = ('Red', 'Green', 'Blue')

    for i, (col, name) in enumerate(zip(color_channels, channel_names)):
        if img_array.ndim >= 3 and img_array.shape[2] > i:
            hist, bins = np.histogram(img_array[:, :, i].flatten(), bins=256, range=[0, 256])
            plt.plot(bins[:-1], hist, color=col, label=f'{name}')
    
    plt.legend()
    plt.xlim([0, 256])

# --------------------------------------------------------------------------

def process_and_display_random_images(directory_path, num_to_display=5):
    """
    Selects and processes a random sample of images from a directory.

    Args:
        directory_path (str): The path to the directory containing images.
        num_to_display (int): The number of random images to display.
    """
    print(f"--- Processing Random Sample from: {directory_path} ---")
    if not os.path.isdir(directory_path):
        print(f"Error: Directory not found at {directory_path}")
        return
        
    # 1. Get a list of all JPG/JPEG files in the directory
    all_files = [f for f in os.listdir(directory_path) 
                 if f.lower().endswith(('.jpg', '.jpeg'))]
    
    if not all_files:
        print("No JPG/JPEG images found in the directory.")
        return

    # 2. Select a random sample of files
    # Ensure we don't try to sample more than the total number of files
    sample_size = min(num_to_display, len(all_files))
    random_files = random.sample(all_files, sample_size)
    
    print(f"Selected {sample_size} random images for processing.")

    # 3. Process and display the sampled files
    for filename in random_files:
        file_path = os.path.join(directory_path, filename)
        
        img, img_array = get_and_process_image(file_path)
        
        if img is not None:
            # Create a figure with two subplots (image and histogram)
            plt.figure(figsize=(14, 5))
            
            # Call the display function
            display_image_with_pixels(img, filename, 121) # 1 row, 2 cols, 1st plot
            
            # Call the histogram function
            plot_color_histogram(img_array, 122) # 1 row, 2 cols, 2nd plot
            
            plt.tight_layout()
            plt.show()
            
    print("-" * 40)

# --------------------------------------------------------------------------
## Example Usage:

# Assuming you have a directory named 'my_image_data'
process_and_display_random_images(file_path, num_to_display=3)
```

```python
# Tranlsate the image to np array
def process_and_display_images(file_path, n_images=1):
    """
    Opens, processes and displays n images from the specified directory.
    Displays image with its dimensions and corresponding color histogram.
    
    Parameters:
    file_path : str
    Path to the directory with images
    n_images : int
    Number of images to process and display

    Returns:
    list of tuples [(PIL.Image, np.array), ...]
    """
    image_files = [
        os.path.join(file_path, f) for f in os.listdir(file_path) if f.lower().endswith('.jpg')
    ][:n_images]

    if not image_files:
        print('No image files found.')
        return[]
    
    # 2 columns per image: image + color histogram
    fig = plt.figure(figsize=(12,6 * len(image_files)))
    results = []

    for idx, img_path in enumerate(image_files):
        try:
            # Load image
            img = Image.open(img_path)
            img_array = np.array(img)
            results.append((img, img_array))

            # Calculate subplot positions
            img_pos = (len(image_files), 2, idx * 2 + 1)
            hist_pos = (len(image_files), 2, idx * 2 + 1)

            # Display image with dimensions
            width, height = img.size
            plt.subplot(*img_pos)
            plt.imshow(img)
            plt.title(f"Image: {os.path.basename(img_path)}")
            plt.axis('off')
            plt.xlabel(f"{width}x{height} pixels")

            # Plot color histogram
            plt.subplot(*hist_pos)
            plt.title('Color Histogram')
            plt.xlabel('Pixel Intensity Values (0-255)')
            plt.ylabel('Frequency (Pixel Count)')
                       
            color_channels = ('r', 'g', 'b')
            channel_names = ('red', 'green', 'blue')

            for i, (col, name) in enumerate(zip(color_channels, channel_names)):
                if img_array.ndim >= 3 and img_array.shape[2] > i:
                    hist, bins = np.histogram(
                        img_array[:, :, i].flatten(),
                        bins=256,
                        range=[0, 255]
                    )
                    plt.plot(bins[:-1], hist, color=col, label=name)

            plt.legend()
            plt.xlim([0, 256])

        except Exception as e:
            print(f"Error porcessin {img_path: {e}}")
            continue

        plt.tight_layout()
        plt.show()

        return results
    
                             

results_santa = process_and_display_images('./data/test/santa')
results_not_santa = process_and_display_images('./data/test/not-a-santa')

```
